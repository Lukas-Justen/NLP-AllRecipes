```python
import pandas as pd
```

```python
data = pd.read_csv("../files/ingredient_data.csv")
```

```python
data
```

```python
def make_lowercase(text):
    return str(text).lower()
```

```python
units = pd.DataFrame(data["unit"].unique(),columns=["unit"])
```

```python
units["unit"] = units["unit"].apply(make_lowercase)
```

```python
names = pd.DataFrame(data["name"].unique(),columns=["name"])
```

```python
names["name"] = names["name"].apply(make_lowercase)
```

```python
comments = pd.DataFrame(data["comment"].unique(),columns=["comment"])
```

```python
comments["comment"] = comments["comment"].apply(make_lowercase)
```

```python
units.to_csv("./units.csv")
```

```python
names.to_csv("./names.csv")
```

```python
comments.to_csv("./comments.csv")
```
